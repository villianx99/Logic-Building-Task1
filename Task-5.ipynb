```python
# Problem 1
def analyze_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like

    # Conditional check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example Usage
likes = [200, 300, 150, 400]
analyze_engagement(likes)
```

```python
# Problem 3
def check_rainfall(rainfall_data, required_level):
    total = 0

    # Calculate total rainfall using loop
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    print("Average Rainfall:", average)

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example Usage
seasonal_data = [70, 75, 68, 80, 67]
required = 70

check_rainfall(seasonal_data, required)
```

```python
# Problem 4
def detect_duplicates(usernames):

    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example Usage
users = ["mayrix", "john", "sara", "mayrix"]
detect_duplicates(users)
```

```python
# Problem 6
def filter_premium_crops(prices):

    premium = []

    # Filter prices greater than 2000
    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


# Example
filter_premium_crops([1500, 2500, 1800, 3200])
```

```python
# Problem 7
def check_application_health(error_count):

    print("Error Count:", error_count)

    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example
check_application_health(7)
```

```python
# Problem 9
def check_attendance(attendance_list):

    total_days = len(attendance_list)
    present_days = 0

    # Count Present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    attendance_percentage = (present_days / total_days) * 100

    print("Attendance Percentage:", attendance_percentage)

    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example
attendance = ["P", "P", "A", "P", "P"]
check_attendance(attendance)
```

```python
# Problem 10
def calculate_electricity_bill(units):

    bill = 0

    # Billing Slab Calculation
    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)


    # Usage Classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example
units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
