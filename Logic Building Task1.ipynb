```python
marks =[45,78,90,33,60]
pass_count=0
fail_count=0
for mark in marks:
 if mark >=50:
   pass_count +=1
 else:
   fail_count +=1
print("total student:",len(marks))
print("passed student:",pass_count)
print("failed student:",fail_count)
```

```python
names =["Alice","bob","CHARLIE"]
cleaned_names=[]
for name in names :
  cleaned =name.strip().lower()
  cleaned_names.append(cleaned)
  print("original List:",names)
  print("cleaned List:",cleaned_names)
```
