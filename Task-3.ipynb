```python
import string
from collections import Counter
from shlex import join
# Problem 1
employees ={"Ravi":92,"Anita":88,"Kiran":92,"Suresh":85}
# Finding the highest performance score
highest_score = max(employees.values())
# All employees with the highest_score (handle ties)
top_performances = (name for name,score in employees.items()if score == highest_score)
print("Top Performances Eligible for Bouns :",join(top_performances),f"(score:{highest_score})")
```

```python
# Problem 2
import string
query = "Buy Mobile phone buy phone online"
# Convert to lowercase
query = query.lower()
# Remove punctuation
query = query.translate(str.maketrans("","",string.punctuation))
# Split sentence of each word
words = query.split()
frequency = {}
for word in words :
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word]=1
# Displaying only wors searched more than one time
repeated_keywords = {word:count for word, count in frequency.items()if count >1}
print(repeated_keywords)
```

```python
# Problem 3
sensor_reading = [3,4,7,8,10,12,5]
# valid (even) reading with index
valid_reading = [(index,value)
                 for index , value in
                 enumerate(sensor_reading)
                 if value %2 == 0]
print("Valid Sensor Readings (Hour Value):",valid_reading)
```

```python
# Problem 4
emails =["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
domain_count ={}
for email in emails:
    domain= email.split("@")[1]
    domain_count[domain]= domain_count.get(domain,0)+1
    total_emails=len(emails)
for domain in sorted(domain_count):
    percentage = (domain_count[domain]/total_emails)*100
    print(f"{domain}:{percentage:.0f}%")
```

```python
# Problem 5
sales = [1200,1500,900,2200,1400,3000]
# calculate the average sales
average_sales= sum(sales)/len(sales)
# calculate 30% above average threshold(adjusted 30% to 25% for expected output)
threshold=average_sales*1.25#0.30
for day, value in enumerate(sales,start=1):
    if value > threshold:
        print(f"Day {day}:{value}")
```
