```python
# Problem 1
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    # Check minimum recharge amount
    if amount < 50:
        return "Recharge amount must be at least ₹50."

    # Check if plan is valid
    if amount in valid_plans:
        return "Recharge Successful "
    else:
        return "Invalid Plan Choose from 199, 299, 399, 599."


# While loop to allow retry
while True:
    try:
        user_amount = int(input("Enter recharge amount: "))
        result = validate_recharge(user_amount)
        print(result)

        # If recharge successful, break loop
        if result == "Recharge Successful ":
            break
        else:
            print("Please try again.\n")

    except ValueError:
        print("Invalid input! Please enter numeric value.\n")
```

```python
# Problem 2
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product} Reorder Alert (Stock: {quantity})")
        else:
            print(f"{product} Stock OK  (Stock: {quantity})")


# Sample Data
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}

# Function Call
check_inventory(inventory)
```

```python
# Problem 3
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


student_marks = [60, 55, 70, 45, 50]

result = process_result(student_marks)
print("Result:", result)
```

```python
# Problem 4
def calculate_fare(distance, peak_hour=False):
    base_fare = 50
    per_km_rate = 12

    # Normal fare calculation
    fare = base_fare + (distance * per_km_rate)

    # Peak hour surcharge (25% extra)
    if peak_hour:
        fare = fare * 1.25

    return fare


# While loop to allow retry
while True:
    try:
        # User input for distance
        distance = float(input("Enter distance travelled (in km): "))

        # User input for peak hour
        peak_input = input("Is it peak hour? (yes/no): ").strip().lower()
        peak_hour = True if peak_input == "yes" else False

        # Calculate fare
        total_fare = calculate_fare(distance, peak_hour)
        print(f"Estimated Cab Fare: ₹{total_fare:.2f}")

        # Break loop after successful calculation
        break

    except ValueError:
        print("Invalid input! Please enter numeric value for distance.\n")
```

```python
# Problem 5
def check_eligibility(attendance):
    present_count =0

    for day in attendance:
        if day == "P":
            present_count += 1
    total_days=len(attendance)
    percentage = (present_count/total_days)*100
    print("Attendance Percentage :", percentage,"%")
    if percentage >= 75 :
        return "Eligible"
    else:
        return "Not eligible"
# Example attendance list
attendance_list=["P","P",'P',"A","P","A","P","P"]
result = check_eligibility(attendance_list)
print(result)
```

```python
# Problem 6
def check_password_strength(password):

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include at least one digit and one special character @#$)"


# Example
password = input("Enter your password: ")
result = check_password_strength(password)
print(result)
```
