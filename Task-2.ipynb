```python
# Problem 1
sentence = input("Enter s sentence :")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("unique words:", unique_words)
```

```python
# Problem  2
employees = {"Ravi": 75000, "Anita": 68000, "Kiran": 72000}
# Highest salary
highest_employee = max(employees, key=employees.get)
print("Highest salary:", highest_employee, "-", employees[highest_employee])
```

```python
#problem 3
numbers = [45, 22, 89, 10, 66]
# Initialize max and min
maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", max(numbers))
print("min:", min(numbers))
```

```python
# Problem 4
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

```python
# Problem 5
attendence = ["P","P","A","P","P"]
present_count = 0

for day in attendence :
    if day == "P":
        present_count += 1
# Precentage calculate
percentage = (present_count/len(attendence))*100
print("attendance percentage:",percentage)
```

```python
# Problem 6
phone_numbers = (9876543210, 9123456789, 9876543210)
# using set removing duploicates
unique_numbers = set(phone_numbers)
print(" Unique phone numbers :", unique_numbers)
```

```python
# Problem 7
text = input("Enter a string:")
char_count = {}
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
print(char_count)
```
